```python
from bing_image_downloader import downloader
from os import path
```

```python
class_names = ['3.5mm', 'Ethernet', 'HDMI', 'USB-A', 'Wall-Power']
search_strings = ['3.5 mm jack', 'Ethernet cord', 'HDMI cord', 'USB cord', 'Wall power cord']

for search_string, class_name in zip(search_strings, class_names):
    downloader.download(search_string, limit=100,  output_dir=path.join('TestDataset','Train', class_name), adult_filter_off=True, force_replace=False, timeout=60)
```
